## Guided 1 (Graph Kosong) - Undirected Graph

```
import matplotlib.pyplot as plt
import networkx as nx

# Posisi node yang digunakan pada modul ini
pos = {
    'A': [0.93888004, -0.0449161],
    'B': [-0.73226196, -0.53345128],
    'C': [0.29334637, 1.0],
    'D': [-0.79991428, 0.56134392],
    'E': [0.29994983, -0.98297654]
}

# Fungsi pendukung untuk mencetak graf
def show_graph(G, pos=None, title=''):
    if pos is None:
        pos = nx.spring_layout(G)

    nx.draw(
        G,
        pos,
        with_labels=True,
        node_color='red',
        node_size=2000,
        font_color='white',
        font_weight='bold',
        width=5
    )

    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(
        G,
        pos,
        edge_labels=edge_labels,
        font_color='blue',
        font_weight='bold',
        font_size=12
    )

    plt.margins(0.2)
    plt.title(title)
    plt.show()

G_undirected = nx.Graph()
# Cetak graf
show_graph(G_undirected,
title="Undirected Graph Kosong")
```

## Guided 2 (DiGraph Kosong) - Directed Graph

```
import matplotlib.pyplot as plt
import networkx as nx

# Posisi node yang digunakan pada modul ini
pos = {
    'A': [0.93888004, -0.0449161],
    'B': [-0.73226196, -0.53345128],
    'C': [0.29334637, 1.0],
    'D': [-0.79991428, 0.56134392],
    'E': [0.29994983, -0.98297654]
}

# Fungsi pendukung untuk mencetak graf
def show_graph(G, pos=None, title=''):
    if pos is None:
        pos = nx.spring_layout(G)

    nx.draw(
        G,
        pos,
        with_labels=True,
        node_color='red',
        node_size=2000,
        font_color='white',
        font_weight='bold',
        width=5
    )

    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(
        G,
        pos,
        edge_labels=edge_labels,
        font_color='blue',
        font_weight='bold',
        font_size=12
    )

    plt.margins(0.2)
    plt.title(title)
    plt.show()

G_directed = nx.DiGraph()
# Cetak graf
show_graph(G_directed,
title="Directed Graph Kosong")
```

## Guided 3 (Menambahkan Node Undirected Graph)

```
# Membuat graf tidak berarah
G_undirected = nx.Graph()

# Cara add_node()
G_undirected.add_node('A')
G_undirected.add_node('B')
G_undirected.add_node('C')
G_undirected.add_node('D')
G_undirected.add_node('E')

# Cara add_nodes_from()
node_data = ['A', 'B', 'C', 'D', 'E']
G_undirected.add_nodes_from(node_data)

# Mencatat posisi node agar konsisten
pos = nx.spring_layout(G_undirected)

# Cetak graf
show_graph(G_undirected, pos=pos, title="Add Node Undirected Graph")

# Melihat jumlah node pada graf
print("Jumlah node:", G_undirected.number_of_nodes())
```

## Guided 4 (Menambahkan Node Directed Graph)

```
# Membuat graf berarah
G_directed = nx.DiGraph()

# Menambahkan node satu per satu
G_directed.add_node('A')
G_directed.add_node('B')
G_directed.add_node('C')
G_directed.add_node('D')
G_directed.add_node('E')

# Atau menambahkan node dalam satu langkah
node_data = ['A', 'B', 'C', 'D', 'E']
G_directed.add_nodes_from(node_data)

# Cetak graf
show_graph(G_directed, pos=pos, title="Add Node Directed Graph")

# Melihat jumlah node pada graf
print("Jumlah node:", G_directed.number_of_nodes())
```

## Guided 5 (Menambahkan edge pada undirected graph)

```
# Cara add_edge()
G_undirected.add_edge('A', 'B')
G_undirected.add_edge('B', 'C')
G_undirected.add_edge('C', 'D')
G_undirected.add_edge('D', 'B')
G_undirected.add_edge('E', 'D')

# Cara add_edges_from()
edge_data = [('A', 'B'), ('B', 'C'),
             ('C', 'D'), ('D', 'B'), ('E', 'D')]
G_undirected.add_edges_from(edge_data)

# Cetak graf
show_graph(G_undirected, pos=pos, title="Add Edge Undirected Graph")

# Melihat jumlah edge pada graf
print("Jumlah edge:", G_undirected.number_of_edges())
```

## Guided 6 (Menambahkan edge pada directed graph)

```
# Cara add_edge()
G_directed.add_edge('A', 'B')
G_directed.add_edge('B', 'C')
G_directed.add_edge('C', 'D')
G_directed.add_edge('D', 'B')
G_directed.add_edge('E', 'D')

# Cara add_edges_from()
edge_data = [('A', 'B'), ('B', 'C'),
             ('C', 'D'), ('D', 'B'), ('E', 'D')]
G_directed.add_edges_from(edge_data)

# Cetak graf
show_graph(G_directed, pos=pos, title="Add Edge Directed Graph")

# Melihat jumlah edge pada graf
print("Jumlah edge:", G_directed.number_of_edges())
```

## Guided 10 (Menambahkan weight pada directed graph)

```
# Cara add_edge() dengan bobot (weight)
G_directed.add_edge('A', 'B', weight=10)
G_directed.add_edge('B', 'C', weight=7)
G_directed.add_edge('C', 'D', weight=4)
G_directed.add_edge('D', 'B', weight=8)
G_directed.add_edge('E', 'D', weight=3)

# Cara add_weighted_edges_from()
edge_weight_data = [('A', 'B', 10),
                    ('B', 'C', 7),
                    ('C', 'D', 4),
                    ('D', 'B', 8),
                    ('E', 'D', 3)]
G_directed.add_weighted_edges_from(edge_weight_data)

# Cetak graf
show_graph(G_directed, pos=pos, title="Add Weight Directed Graph")
```

## Unguided 1


```
import networkx as nx
import matplotlib.pyplot as plt

# Buat graf berbobot tak berarah (undirected graph)
G = nx.Graph()

# Tambahkan simpul (node)
kota = ["Oradea", "Zerind", "Arad", "Timisoara", "Lugoj", "Mehadia", "Drobeta", "Craiova", "Rimnicu Vilcea", "Sibiu", "Fagaras", "Pitesti", "Bucharest", "Giurgiu", "Urziceni", "Hirsova", "Eforie", "Vaslui", "Iasi", "Neamt"]
G.add_nodes_from(kota)

# Tambahkan sisi (edge) dengan bobot sesuai peta
edges = [
    ("Oradea", "Zerind", 71), ("Oradea", "Sibiu", 151),
    ("Zerind", "Arad", 75), ("Arad", "Sibiu", 140), ("Arad", "Timisoara", 118),
    ("Timisoara", "Lugoj", 111), ("Lugoj", "Mehadia", 70),
    ("Mehadia", "Drobeta", 75), ("Drobeta", "Craiova", 120),
    ("Craiova", "Rimnicu Vilcea", 146), ("Craiova", "Pitesti", 138),
    ("Rimnicu Vilcea", "Sibiu", 80), ("Rimnicu Vilcea", "Pitesti", 97),
    ("Sibiu", "Fagaras", 99), ("Fagaras", "Bucharest", 211),
    ("Pitesti", "Bucharest", 101), ("Bucharest", "Giurgiu", 90),
    ("Bucharest", "Urziceni", 85), ("Urziceni", "Vaslui", 142),
    ("Vaslui", "Iasi", 92), ("Iasi", "Neamt", 87),
    ("Urziceni", "Hirsova", 98), ("Hirsova", "Eforie", 86)
]

G.add_weighted_edges_from(edges)

# Gambar graf
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, seed=42)  # Tata letak graf
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=10)
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)
plt.title("Weighted Undirected Graph - European Cities")
plt.show()

# Menampilkan shortest path untuk semua pasangan simpul
shortest_paths = dict(nx.all_pairs_dijkstra_path(G))
shortest_distances = dict(nx.all_pairs_dijkstra_path_length(G))

# Contoh menampilkan shortest path dari "Arad" ke "Bucharest"
print("Shortest path from Arad to Bucharest:", shortest_paths["Arad"]["Bucharest"])
print("Total distance:", shortest_distances["Arad"]["Bucharest"])
```

```
import networkx as nx
import matplotlib.pyplot as plt

# Buat graf berbobot
G = nx.Graph()

# Tambahkan simpul (nodes)
kota = ["Jakarta", "Cirebon", "Semarang", "Surabaya", "Bandung", "Yogyakarta", "Surakarta", "Malang"]
G.add_nodes_from(kota)

# Tambahkan sisi (edges) dengan bobot (weight)
edges = [
    ("Jakarta", "Cirebon", 327),
    ("Jakarta", "Bandung", 270),
    ("Cirebon", "Semarang", 305),
    ("Cirebon", "Bandung", 120),
    ("Cirebon", "Yogyakarta", 210),
    ("Semarang", "Yogyakarta", 109),
    ("Semarang", "Surakarta", 97),
    ("Surakarta", "Yogyakarta", 60),
    ("Surabaya", "Malang", 94),
    ("Surabaya", "Surakarta", 370),
    ("Surabaya", "Semarang", 369),
    ("Bandung", "Yogyakarta", 373)
]

G.add_weighted_edges_from(edges)

# Gambar graf
pos = nx.spring_layout(G)  # Tata letak untuk graf
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=3000, font_size=10)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Graf Berbobot Kota-Kota di Pulau Jawa")
plt.show()
```
