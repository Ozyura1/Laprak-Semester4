## Guided 1 (Penggunaan Figure dalam Matplotlib)

```python
import matplotlib.pyplot as plt

# Inisialisasi Figure dengan ukuran 8x6
fig = plt.figure(figsize=(8,6))

# Membuat subplot dalam figure
ax = fig.add_subplot(1,1,1)  # 1 baris, 1 kolom, indeks 1

# Membuat plot sederhana
ax.plot([1,2,3,4], [10,20,25,30], label='Data1')

# Menambah label dan judul
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_title("Contoh Plot dalam Figure")

# Menampilkan legenda
ax.legend()

# Menampilkan plot
plt.show()
```

program diatas berfungsi untuk membuat sebuah plot dalam figure dimana menggunakan Matplotlib untuk membuat dan menampilkan plot sederhana. program menggunakan Figure ukuran 8x6 dan menambahkan subplot tunggal kedalam, kemudian program menggambar grafik garis berdasarkan dua nilai yaitu X dan Y. terakhir grafik ditampilkan menggunakan plt.show()

## Guided 3 (Memberikan Anotasi Pada Label)

```python
import matplotlib.pyplot as plt

# Membuat figure
fig = plt.figure()

# Membuat line chart
x = [1, 2, 3, 4, 5, 6]
y1 = [1, 2, 3, 4, 5, 6]
y2 = [2, 3, 5, 6, 7, 8]

plt.plot(x, y1, label="Data 1")  # Garis pertama
plt.plot(x, y2, label="Data 2")  # Garis kedua

# Menambahkan anotasi pada titik (x=3, y=5)
plt.annotate("Poin Penting", xy=(3, 5), xytext=(3, 6),
             arrowprops=dict(facecolor='pink', shrink=0.05))

# Menampilkan legenda berdasarkan label pada data
plt.legend()

# Menampilkan figure
plt.show()
```

## Guided 5 (Scatter Plot)

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5, 6, 7, 8]
y1 = [10, 21, 25, 30, 34, 41, 47, 50]
y2 = [12, 20, 26, 29, 33, 37, 41, 43]

# Membuat figure dengan ukuran 10x5 inch
fig = plt.figure(figsize=(10,5))

# Membuat scatter plot
plt.scatter(x, y1, s=100, c='pink', alpha=0.65, label='Data 1')  # Titik merah, ukuran besar
plt.scatter(x, y2, s=50, c='green', alpha=0.65, label='Data 2')  # Titik biru, ukuran kecil

# Labeling plot
plt.title("Contoh Scatter Plot")  # Menambahkan judul
plt.xlabel("Data X")  # Label untuk sumbu X
plt.ylabel("Data Y")  # Label untuk sumbu Y

# Menampilkan legenda berdasarkan label pada data
plt.legend()

# Menampilkan figure
plt.show()
```

## Guided 7 (Fill Between)

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]
y1 = [9, 20, 27, 35]
y_main = [7, 18, 22, 29]  # Data garis utama
y2 = [6, 15, 18, 23]

# Membuat figure
plt.figure(figsize=(10, 6))

# Membuat fill_between plot
plt.fill_between(x, y1, y2, color='pink', alpha=0.5, label='Fill Plot')

# Membuat plot garis utama
plt.plot(x, y_main, color='red', linewidth=2, label='Garis Utama')

# Labeling plot
plt.title("Fill Between")
plt.xlabel("X")
plt.ylabel("Y")

# Menampilkan legend dan figure
plt.legend(loc='lower right') # legenda di kanan bawah
plt.show()
```

program diatas membuat fill between yang berfungsi untuk menyoroti area diantara dua garis dengan warna. figure dibuat dengan ukuran 10x6. pada area y1 dan y2 diwarnai warna pink dengan tranparasi alpha = 0.5 menggunakan plt.fill_between(), sedangkan y_main ditampilkan dengan warna garis erah dengan ketebalan 2 menggunakan plt.plot() dan ditambahkan tampilkan menggunakaan plt.show()

## Guided 8 (Stack Plot)

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 10, 15, 20]
y3 = [2, 5, 8, 12]

# Membuat figure
fig = plt.figure(figsize=(10,6))

# Membuat stack plot
plt.stackplot(x, y1, y2, y3, labels=['Y1', 'Y2', 'Y3'])

# Labeling plot
plt.title("Stack Plot")
plt.xlabel("X")
plt.ylabel("Y")

# Menampilkan legend dan figure
plt.legend()
plt.show()
```

program diatas berfungsi membuat stack plot grafik bertumpuk yang menunjukan dari beberapa data set. figure dibuat dengan ukuran 10x6 untuk wadah grafik. pada grafik are bertumpuk dibuat menggunakan plt.stackplot(), dengan masing masing diberi nilai Y1,Y2,Y3, dan ditambahkan tampilan output seperti pada output diatas

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data mahasiswa berdasarkan tabel pada gambar
data = {
    "Nama": ["Budi", "Siti", "Asep", "Dewi", "Andi", "Lestari", "Joko", "Rina", "Fajar", "Indah",
             "Yuni", "Bambang", "Ayu", "Fitri", "Rudi", "Maya", "Hadi", "Sri", "Wawan"],
    "Algoritma Pemrograman": [81.24, 98.52, 91.96, 87.96, 74.68, 74.68, 71.74, 95.99, 88.03, 91.24,
                              85.74, 82.96, 78.74, 85.72, 94.97, 76.37, 82.96, 75.5, 83.16],
    "Matematika Diskrit": [73.66, 84.86, 71.03, 97.28, 77.76, 89.88, 79.35, 95.88, 86.4, 75.55,
                           75.88, 71.36, 79.76, 82.35, 98.18, 77.87, 84.94, 76.07, 79.68],
    "Dasar Kecerdasan Artifisial": [95.89, 88.7, 79.93, 71.91, 76.79, 79.76, 91.89, 78.15, 96.62, 84.17,
                                    82.83, 78.13, 76.23, 73.59, 81.72, 81.84, 79.21, 78.24, 83.16],
    "Tahun Angkatan": [2023, 2024, 2022, 2024, 2023, 2022, 2022, 2022, 2023, 2023,
                       2022, 2023, 2022, 2022, 2022, 2023, 2022, 2023, 2022]
}

# Konversi ke DataFrame
df = pd.DataFrame(data)

# Mengelompokkan berdasarkan Tahun Angkatan dan menghitung statistik
grouped = df.groupby("Tahun Angkatan")[["Algoritma Pemrograman", "Matematika Diskrit", "Dasar Kecerdasan Artifisial"]]

# Hitung nilai maksimum, minimum, dan rata-rata
nilai_max = grouped.max()
nilai_min = grouped.min()
nilai_mean = grouped.mean()

# Plot hasil
plt.figure(figsize=(10, 5))
plt.plot(nilai_max.mean(axis=1), marker='o', linestyle='-', label='Maksimum', color='red')
plt.plot(nilai_min.mean(axis=1), marker='s', linestyle='--', label='Minimum', color='blue')
plt.plot(nilai_mean.mean(axis=1), marker='^', linestyle='-', label='Rata-rata', color='green')

# Labeling
plt.title("Grafik Nilai Maksimum, Minimum, dan Rata-rata per Angkatan")
plt.xlabel("Tahun Angkatan")
plt.ylabel("Nilai")
plt.legend()
plt.grid(True)

# Tampilkan plot
plt.show()
```

##Unguided2

```python
import networkx as nx
import matplotlib.pyplot as plt

# Membuat graf pertama: Kota-kota besar di Eropa
G1 = nx.Graph()

edges_europe = [
    ("Oradea", "Zerind", 71), ("Zerind", "Arad", 75),
    ("Arad", "Timisoara", 118), ("Arad", "Sibiu", 140),
    ("Timisoara", "Lugoj", 111), ("Lugoj", "Mehadia", 70),
    ("Mehadia", "Drobeta", 75), ("Drobeta", "Craiova", 120),
    ("Craiova", "Pitesti", 138), ("Pitesti", "Bucharest", 101),
    ("Sibiu", "Fagaras", 99), ("Fagaras", "Bucharest", 211),
    ("Bucharest", "Giurgiu", 90), ("Sibiu", "Rimnicu Vilcea", 80),
    ("Rimnicu Vilcea", "Pitesti", 97), ("Craiova", "Rimnicu Vilcea", 146)
]

for edge in edges_europe:
    G1.add_edge(edge[0], edge[1], weight=edge[2])

# Posisi node untuk graf Eropa agar mirip dengan gambar soal
pos_europe = {
    "Oradea": (0, 5), "Zerind": (1, 5), "Arad": (2, 5),
    "Timisoara": (3, 4), "Lugoj": (4, 4), "Mehadia": (5, 4),
    "Drobeta": (6, 3), "Craiova": (7, 3), "Pitesti": (8, 3),
    "Bucharest": (9, 3), "Giurgiu": (10, 2), "Fagaras": (8, 4),
    "Sibiu": (3, 5), "Rimnicu Vilcea": (6, 4)
}

# Membuat graf kedua: Kota-kota besar di Indonesia (Pulau Jawa)
G2 = nx.Graph()

edges_indonesia = [
    ("Jakarta", "Cirebon", 327), ("Jakarta", "Bandung", 270),
    ("Cirebon", "Semarang", 350), ("Bandung", "Yogyakarta", 390),
    ("Semarang", "Yogyakarta", 210), ("Semarang", "Surabaya", 350),
    ("Yogyakarta", "Surabaya", 320), ("Surabaya", "Malang", 98)
]

for edge in edges_indonesia:
    G2.add_edge(edge[0], edge[1], weight=edge[2])

# Posisi node untuk graf Indonesia agar mirip dengan gambar soal
pos_indonesia = {
    "Jakarta": (0, 3), "Bandung": (-1, 2), "Cirebon": (1, 3),
    "Semarang": (2, 2), "Yogyakarta": (1, 1), "Surabaya": (3, 2),
    "Malang": (3, 1)
}

# Membuat subplot untuk menampilkan kedua graf dalam satu tampilan
plt.figure(figsize=(12, 6))

# Plot graf Eropa
plt.subplot(1, 2, 1)
nx.draw(G1, pos_europe, with_labels=True, node_color="lightblue", node_size=2500, font_size=8, edge_color="gray")
labels1 = nx.get_edge_attributes(G1, 'weight')
nx.draw_networkx_edge_labels(G1, pos_europe, edge_labels=labels1, font_size=7)
plt.title("Graf Kota-Kota di Eropa")

# Plot graf Indonesia
plt.subplot(1, 2, 2)
nx.draw(G2, pos_indonesia, with_labels=True, node_color="lightgreen", node_size=2500, font_size=10, edge_color="gray")
labels2 = nx.get_edge_attributes(G2, 'weight')
nx.draw_networkx_edge_labels(G2, pos_indonesia, edge_labels=labels2, font_size=7)
plt.title("Graf Kota-Kota di Pulau Jawa")

plt.tight_layout()
plt.show()

# Mencari jalur terpendek menggunakan Dijkstra untuk kedua graf
def shortest_paths(G):
    return dict(nx.all_pairs_dijkstra_path_length(G))

print("\nJalur terpendek antar semua pasangan kota di Eropa:")
shortest_paths_europe = shortest_paths(G1)
for kota_asal, paths in shortest_paths_europe.items():
    for kota_tujuan, jarak in paths.items():
        if kota_asal != kota_tujuan:
            print(f"Jarak dari {kota_asal} ke {kota_tujuan}: {jarak} km")

print("\nJalur terpendek antar semua pasangan kota di Indonesia:")
shortest_paths_indonesia = shortest_paths(G2)
for kota_asal, paths in shortest_paths_indonesia.items():
    for kota_tujuan, jarak in paths.items():
        if kota_asal != kota_tujuan:
            print(f"Jarak dari {kota_asal} ke {kota_tujuan}: {jarak} km")
```

Program ini berfungsi untuk membuat dan menganalisis dua graf berbobot menggunakan library NetworkX dan Matplotlib. Graf pertama merepresentasikan jaringan kota-kota besar di Eropa, dan graf kedua menggambarkan kota-kota besar di Pulau Jawa, Indonesia. Setiap simpul dalam graf mewakili sebuah kota, dan sisi antar simpul memiliki bobot yang menunjukkan jarak antar kota. Program ini menampilkan kedua graf dalam satu figure dengan posisi simpul yang telah diatur agar menyerupai gambar. Selain itu, program menggunakan algoritma Dijkstra untuk mencari jalur terpendek antara semua pasangan kota dalam kedua graf dan mencetak hasilnya di terminal.
